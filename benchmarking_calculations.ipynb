```python
import pandas as pd

# Load the CSV
df = pd.read_csv('voting_rule_average_performance_4.csv')
df.columns = df.columns.str.lower()

# Map domain IDs to names
domain_mapping = {
    1.0: 'Geography',
    2.0: 'Movies',
    3.0: 'Paintings'
}
df['domain'] = df['domain'].map(domain_mapping)

# Filter relevant columns
df = df[['voting rule', 'domain', 'elicitation format', 
         'kendall tau vote (mean)', 'kendall tau sp (mean)']].copy()

# ------------------ Table 1: Average across domains ------------------ #
table1 = df.groupby(['elicitation format', 'voting rule'])[
    ['kendall tau vote (mean)', 'kendall tau sp (mean)']
].mean().unstack('voting rule').round(2)

# Flatten multi-index columns
table1.columns = [f'{rule} {metric.split()[-2]}' for metric, rule in table1.columns]
table1 = table1.reset_index()

# Save Table 1
#table1.to_csv('kendall_tau_avg_across_domains_subset_4.csv', index=False)

print(table1)

# ------------------ Table 2: Average per domain ------------------ #
table2 = df.groupby(['elicitation format', 'domain'])[
    ['kendall tau vote (mean)', 'kendall tau sp (mean)']
].mean().round(2).reset_index()

# Pivot the table
table2_pivot = table2.pivot(index='elicitation format', columns='domain')
table2_pivot.columns = [f'{domain} {metric.split()[-2]}' for metric, domain in table2_pivot.columns]
table2_pivot = table2_pivot.reset_index()

# Save Table 2
#table2_pivot.to_csv('kendall_tau_avg_per_domain_subset_4.csv', index=False)

print(table2_pivot)
```
